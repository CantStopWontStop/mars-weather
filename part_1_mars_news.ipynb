````python
# Import Splinter and BeautifulSoup
from splinter import Browser
from bs4 import BeautifulSoup
from webdriver_manager.chrome import ChromeDriverManager
from selenium import webdriver
import requests


````

````python
executable_path = {'executable_path': ChromeDriverManager().install()}
browser = Browser('chrome', **executable_path, headless=False)
URL  = "https://redplanetscience.com/"
browser.visit(URL)
html = browser.html
sp   = BeautifulSoup(html,"lxml")
print(sp.prettify())
browser.quit()
````

### Step 2: Scrape the Website

Create a Beautiful Soup object and use it to extract text elements from the website.

````python
# Create a Beautiful Soup 

section = sp.find('section',class_ = "image_and_description_container")
articles = section.find_all('div', class_='row')


articles_dict   = []


for row in articles:
    title_row = row.find('div', class_='content_title')
    prev_row = row.find('div', class_='article_teaser_body')
    
    
    img_row = row.find('div', class_='list_image')
    img = img_row.find('img')
    
    dictionary = {'title': title_row.text,
    'preview': prev_row.text,
    'iage': img['src']}
    articles_dict.append(dictionary)
    
print(articles_dict)
````
